```python
num_shades = 10

# Create a discrete colormap with grey shades
# Create a discrete colormap with grey shades
cmap = ListedColormap(sns.color_palette("Greys", num_shades))
bounds = np.arange(0.5, num_shades + 1.5, 1)  # Create boundaries for discrete colors from 1 to 10
norm = BoundaryNorm(boundaries=bounds, ncolors=cmap.N)
# Create the heatmap with discrete colormap
plt.figure(figsize=(5, 8))
ax = sns.heatmap(cell_counts, cmap=cmap, norm=norm, cbar_kws={'ticks': np.arange(num_shades), 'label': 'number of cells'}, annot=False)

# Customize the heatmap
ax.set_yticks(np.arange(len(cell_counts.index)) + 0.5)  # Center y-ticks
ax.set_yticklabels(['animals\n(NeuroPAL, GCaMP6s)'] * len(cell_counts.index), rotation=0, fontsize=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, fontsize=12, fontweight='bold')

# Adjust the colorbar
cbar = ax.collections[0].colorbar
cbar.set_ticks(np.arange(1, num_shades + 1))  # Discrete ticks at each shade
cbar.set_ticklabels([str(i) for i in range(1, num_shades + 1)])  # Label each tick from 1 to 10
cbar.set_label('number of cells', fontsize=12, rotation=270, labelpad=20)

# Show the plot
plt.show()
```
