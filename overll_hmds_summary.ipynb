# Radii

```python
buffer_dict = buffer.set_index("Filename").iloc[:, 6].apply(lambda x: x.replace(",", " ")).to_dict()
clusters_dict = clusteres.set_index("Worm ID").iloc[:, 6].apply(lambda x: x.replace(",", " ")).to_dict()
whole_stim_dict = all_w.set_index("Worm ID").iloc[:, 6].apply(lambda x: x.replace(",", " ")).to_dict()
unclustered_dict = not_clusters_full.set_index("Worm ID").iloc[:, 6].apply(lambda x: x.replace(",", " ")).to_dict()

# Print the resulting dictionary
```

```python
# Function to process a dictionary into numeric data for a boxplot
def process_dictionary(data_dict):
    all_data = []  # Collect all numeric lists
    for key, value in data_dict.items():
        numeric_values = list(map(float, value.split()))  # Convert space-separated string to list of floats
        all_data.append(numeric_values)  # Append to the list
    return all_data

# Process each dictionary
buffer_data = process_dictionary(buffer_dict)
clusters_data = process_dictionary(clusters_dict)
whole_stim_data = process_dictionary(whole_stim_dict)
unclustered_data = process_dictionary(unclustered_dict)

# Create boxplots for each dictionary
plt.figure(figsize=(12, 8))

# Buffer
plt.subplot(2, 2, 1)
plt.boxplot(buffer_data, labels=buffer_dict.keys(), patch_artist=True, boxprops=dict(facecolor="skyblue"))
plt.title("Buffer")
plt.ylabel("Values")

# Clusters
plt.subplot(2, 2, 2)
plt.boxplot(clusters_data, labels=clusters_dict.keys(), patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title("Clusters")
plt.ylabel("Values")

# Whole Stimulus
plt.subplot(2, 2, 3)
plt.boxplot(whole_stim_data, labels=whole_stim_dict.keys(), patch_artist=True, boxprops=dict(facecolor="lightcoral"))
plt.title("Whole Stimulus")
plt.ylabel("Values")

# Unclustered
plt.subplot(2, 2, 4)
plt.boxplot(unclustered_data, labels=unclustered_dict.keys(), patch_artist=True, boxprops=dict(facecolor="gold"))
plt.title("Unclustered")
plt.ylabel("Values")

plt.tight_layout()
plt.show()
```

```python
combined_data = [
    [value for sublist in whole_stim_data for value in sublist],  # Flatten Whole Stimulus data
    [value for sublist in buffer_data for value in sublist],     # Flatten Buffer data
    [value for sublist in clusters_data for value in sublist],   # Flatten Clusters data
    [value for sublist in unclustered_data for value in sublist] # Flatten Unclustered data
]

# Labels for each category
labels = ["Whole Stimulus", " Whole Buffer", "Stimulus - Clusters", "Stimulus - Unclustered"]

# Create the combined box plot
plt.figure(figsize=(4.4, 4.7))
bp = plt.boxplot(combined_data, labels=labels, patch_artist=True, boxprops=dict(facecolor='#8dd3c7'))

# Rotate the x-tick labels
plt.xticks(range(1, len(labels) + 1), labels, rotation=25, ha='right', fontsize=14)

# Add y-axis label and title
plt.ylabel("Values", fontsize=14)
plt.title("Hyperbolic Radius", fontsize=14)

# Adjust layout and display the plot
plt.tight_layout()
plt.savefig('/Users/iuliarusu/Documents/Sharpee/final_images/figure2/radians.svg', format='svg', bbox_inches='tight')
plt.show()
```

# BIC

```python
cluster_bic = clusteres['BIC'] 
whole_stim_bic = all_w['BIC']
whole_buffer = buffer['BIC'] 
unclustered_bic = not_clusters_full['BIC'] 
# unclustered_under = not_clusters_undersampled['BIC'] 

```

```python
# Combine the data into a reordered dictionary
bic_data = {
    "Whole Stimulus": whole_stim_bic,
    "Whole Buffer": whole_buffer,
    "Stimulus - Clusters": cluster_bic,
    "Stimulus - Unidentified": unclustered_bic,
}
# Compute means and standard deviations
labels = list(bic_data.keys())
means = [np.mean(bic_data[label]) for label in labels]
std_devs = [np.std(bic_data[label]) for label in labels]

# Perform pairwise t-tests between "Whole Stimulus" and the others
reference_group = "Whole Stimulus"
p_values = {}
for label in labels:
    if label != reference_group:
        t_stat, p_val = ttest_ind(bic_data[reference_group], bic_data[label], equal_var=False)
        p_values[label] = p_val

# Define thresholds for star system
def p_to_star(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return "ns"

# Define patterns for each bar
hatch_patterns = ["//", "\\\\", "xx", ""]

# Plotting
x_positions = np.arange(len(labels))
bar_width = 0.6

plt.figure(figsize=(4.4, 4.7))
for i, (label, hatch) in enumerate(zip(labels, hatch_patterns)):
    plt.bar(
        x_positions[i],
        means[i],
        yerr=std_devs[i],
        capsize=5,
        alpha=0.7,
        color='#80b1d3',
        width=bar_width,
        hatch=hatch,  # Add hatch patterns
        edgecolor="black"
    )

# Annotate t-test p-values above the bars with star system
# for i, label in enumerate(labels):
#     if label in p_values:
#         stars = p_to_star(p_values[label])
#         if stars != "ns":  # Only annotate significant values
#             # plt.text(x_positions[i], means[i] + std_devs[i] + 0.05, stars, ha='center', fontsize=14, color='black')

# Customizing the plot

plt.xticks(x_positions, labels, rotation=25, ha='right', fontsize=14)
plt.yticks(np.arange(0,13, 2 ), fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Mean BIC", fontsize=16)
# plt.xlabel("Groups", fontsize=16)
# plt.title("Hyperbolic Dimensionality", fontsize=18)

plt.tight_layout()
plt.savefig('/Users/iuliarusu/Documents/Sharpee/final_images/figure2/all_bic_with_ttest_stars.svg', format='svg', bbox_inches='tight')
plt.show()
```
